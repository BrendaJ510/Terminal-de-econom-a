```python
# Lista de palabras
palabras = ["python", "código", "datos", "python", "análisis", "datos", "código"]

# Crear un diccionario para contar las apariciones de cada palabra
conteo_palabras = {}
for palabra in palabras:
    if palabra in conteo_palabras:
        conteo_palabras[palabra] += 1
    else:
        conteo_palabras[palabra] = 1

# Encontrar la palabra más repetida
palabra_mas_repetida = max(conteo_palabras, key=conteo_palabras.get)

# Imprimir el diccionario resultante
print("Diccionario de conteo de palabras:", conteo_palabras)
print("Palabra más repetida:", palabra_mas_repetida)
```

```python
def busqueda_binaria(lista, objetivo, inicio, fin): # búsqueda en listas ordenadas
    if inicio > fin:
        return -1  # objetivo no encontrado
    else:
        medio = (inicio + fin) // 2
    if lista[medio] == objetivo:
        return medio
    elif lista[medio] < objetivo:
        return busqueda_binaria(lista, objetivo, medio + 1, fin)  # si es más pequeño se actualiza el inicio
    else:
        return busqueda_binaria(lista, objetivo, inicio, medio - 1)  # si es más grande se actualiza el fin

lista = [1, 3, 4, 7, 10, 15, 17, 29, 40]
objetivo = 17
resultado = busqueda_binaria(lista, objetivo, 0, len(lista) - 1)

print("Índice del objetivo:", resultado)
```

```python
def calcular(n1, n2, operacion):
    if operacion == '+':
        return n1 + n2
    elif operacion == '-':
        return n1 - n2
    elif operacion == '*':
        return n1 * n2
    elif operacion == '/':
        if n2 != 0:
            return n1 / n2
        else:
            return "Error: División por cero."
    elif operacion == '**':
        return n1 ** n2
    elif operacion == '%':
        if n2 != 0:
            return n1 % n2
        else:
            return "Error: División por cero."
    else:
        return "Operación no válida."

# Ejemplo de uso
resultado = calcular(10, 5, '+')
print(resultado)  # Salida: 15

```

```python
def calculadora():
    while True:
        try:
            # Solicitar los números al usuario
            num1 = float(input("Introduce el primer número: "))
            num2 = float(input("Introduce el segundo número: "))

            # Solicitar la operación al usuario
            operacion = input("Introduce la operación (+, -, *, /, **, %): ")

            if operacion == '+':
                return num1 + num2
            elif operacion == '-':
                return num1 - num2
            elif operacion == '*':
                return num1 * num2
            elif operacion == '/':
                if num2 == 0:
                    raise ZeroDivisionError("No se puede dividir entre cero.")
                return num1 / num2
            elif operacion == '**':
                return num1 ** num2
            elif operacion == '%':
                return num1 % num2
            else:
                raise ValueError("Operación no válida.")
        except ValueError as e:
            print(f"Error: {e}. Inténtalo de nuevo.")
        except ZeroDivisionError as e:
            print(f"Error: {e}. Inténtalo de nuevo.")

# Ejecutar la función
resultado = calculadora()
print("Resultado:", resultado)

```

```python
def convertir_temperatura():
    try:
        # Solicitar la temperatura y las unidades al usuario
        temperatura = float(input("Introduce la temperatura: "))
        unidad_origen = input("Introduce la unidad de origen (C, F, K): ").upper()
        unidad_destino = input("Introduce la unidad de destino (C, F, K): ").upper()

        # Validar la unidad de origen
        if unidad_origen not in ['C', 'F', 'K']:
            raise ValueError("Unidad de origen no válida.")
        # Validar la unidad de destino
        if unidad_destino not in ['C', 'F', 'K']:
            raise ValueError("Unidad de destino no válida.")
        # Evitar temperaturas Kelvin negativas
        if unidad_origen == 'K' and temperatura < 0:
            raise ValueError("La temperatura en Kelvin no puede ser negativa.")

        # Realizar la conversión
        if unidad_origen == 'C':
            if unidad_destino == 'F':
                resultado = (temperatura * 9/5) + 32
            elif unidad_destino == 'K':
                resultado = temperatura + 273.15
            else:
                resultado = temperatura
        elif unidad_origen == 'F':
            if unidad_destino == 'C':
                resultado = (temperatura - 32) * 5/9
            elif unidad_destino == 'K':
                resultado = (temperatura - 32) * 5/9 + 273.15
            else:
                resultado = temperatura
        elif unidad_origen == 'K':
            if unidad_destino == 'C':
                resultado = temperatura - 273.15
            elif unidad_destino == 'F':
                resultado = (temperatura - 273.15) * 9/5 + 32
            else:
                resultado = temperatura

        return f"La temperatura convertida es: {resultado:.2f} {unidad_destino}"

    except ValueError as e:
        print(f"Error: {e}. Inténtalo de nuevo.")

# Ejecutar la función
resultado = convertir_temperatura()
print(resultado)
```

```python
import random
import string

def generar_contraseña():
    try:
        # Solicitar la longitud de la contraseña al usuario
        longitud = int(input("Introduce la longitud de la contraseña (mínimo 6): "))

        # Verificar que la longitud sea válida
        if longitud < 6:
            raise ValueError("La longitud debe ser al menos 6.")

        # Caracteres permitidos en la contraseña
        caracteres = string.ascii_letters + string.digits + string.punctuation

        # Generar la contraseña aleatoria
        contraseña = ''.join(random.choice(caracteres) for _ in range(longitud))

        return contraseña

    except ValueError as e:
        print(f"Error: {e}. Inténtalo de nuevo.")
        return generar_contraseña()

# Ejecutar la función
contraseña = generar_contraseña()
print("Contraseña generada:", contraseña)
```

```python


def calcular_estadisticas():
    try:
        # Solicitar la lista de números al usuario
        lista = input("Introduce una lista de números separados por comas: ").split(',')

        # Convertir los elementos a números
        lista = [float(elemento) for elemento in lista]

        if not lista:
            raise ValueError("La lista está vacía.")

        # Calcular el promedio
        promedio = sum(lista) / len(lista)

        # Encontrar el número más grande y el más pequeño
        num_max = max(lista)
        num_min = min(lista)

        # Calcular la mediana
        lista_ordenada = sorted(lista)
        n = len(lista_ordenada)
        if n % 2 == 1:
            mediana = lista_ordenada[n // 2]
        else:
            mediana = (lista_ordenada[n // 2 - 1] + lista_ordenada[n // 2]) / 2

        return promedio, num_max, num_min, mediana

    except ValueError as e:
        return f"Error: {e}"

# Ejecutar la función
estadisticas = calcular_estadisticas()
if isinstance(estadisticas, str):
    print(estadisticas)  # Imprimir error si hay
else:
    promedio, num_max, num_min, mediana = estadisticas
    print("Promedio:", promedio)
    print("Número más grande:", num_max)
    print("Número más pequeño:", num_min)
    print("Mediana:", mediana)
```

```python
def validar_usuario_y_contraseña():
    try:
        # Solicitar el nombre de usuario y la contraseña al usuario
        usuario = input("Introduce el nombre de usuario: ")
        contraseña = input("Introduce la contraseña: ")

        # Verificar que las entradas no estén vacías
        if not usuario or not contraseña:
            raise ValueError("El usuario o la contraseña no pueden estar vacíos.")

        # Verificar que el usuario tenga al menos 5 caracteres
        if len(usuario) < 5:
            raise ValueError("El usuario debe tener al menos 5 caracteres.")

        # Verificar que la contraseña tenga al menos 8 caracteres
        if len(contraseña) < 8:
            raise ValueError("La contraseña debe tener al menos 8 caracteres.")

        # Verificar que la contraseña incluya al menos un número y una mayúscula
        if not any(char.isdigit() for char in contraseña):
            raise ValueError("La contraseña debe incluir al menos un número.")
        if not any(char.isupper() for char in contraseña):
            raise ValueError("La contraseña debe incluir al menos una mayúscula.")

        return "Usuario y contraseña válidos."

    except ValueError as e:
        print(f"Error: {e}. Inténtalo de nuevo.")
        return validar_usuario_y_contraseña()

# Ejecutar la función
resultado = validar_usuario_y_contraseña()
print(resultado)
```

```python
# Importamos la librería random
import random

# Definimos el precio inicial
precio_inicial = 100

# Inicializamos la lista de precios con el precio inicial
precios = [precio_inicial]

# Generamos 10 precios posteriores
for _ in range(10):
    # Definimos la variación aleatoria entre -5% y +5%
    variacion = random.uniform(-0.05, 0.05)
    # Calculamos el nuevo precio
    nuevo_precio = precios[-1] * (1 + variacion)
    # Agregamos el nuevo precio a la lista
    precios.append(nuevo_precio)

# Imprimimos la lista de precios
print("Simulación de precios de la acción:")
for precio in precios:
    print(f"${precio:.2f}")
```

```python
import numpy as np

# Definir el vector
v = np.array([3, -4, 1])

# Calcular la norma Euclidiana (L2)
norma_L2 = np.linalg.norm(v)
print(f"Norma Euclidiana (L2): {norma_L2}")

# Calcular la norma Manhattan (L1)
norma_L1 = np.sum(np.abs(v))
print(f"Norma Manhattan (L1): {norma_L1}")

# Calcular la norma Máxima (L∞)
norma_Linf = np.max(np.abs(v))
print(f"Norma Máxima (L∞): {norma_Linf}")
```
